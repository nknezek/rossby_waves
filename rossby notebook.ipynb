Solves the system of equations 
$$\frac{d\vec{u}}{dt}=-\nabla P + 2\Omega \times \vec{u}$$
$$ \nabla \cdot \vec{u} = 0 $$ 
Nk points in radial direction <br>
Nl points in theta direction

```python
import numpy as np
import scipy as sp
import scipy.sparse as sparse
import scipy.sparse.linalg as LA
import matplotlib.pyplot as plt
import matplotlib.pylab as pyl
import matplotlib as mpl
import rossby_functions as rf

from numpy import sin
from numpy import cos
%matplotlib inline
reload(rf)
```

```python
# modes to simulate
m_min = 2
m_max = 2
m_values = range(m_min,m_max+1)
l_values = {}
l_values[m_values[0]]=[m_values[0]]

# Size of grid
Nk = 1
Nl = 31
Nm = 1

model_variables = ('uk','ul','um','p')

# Physical Constants
R = 3160. # Outer core radius in km
Omega = np.pi/(24.0*3600.0) # rotation rate in rad/s
rho = 1.e5  # density in kg/m^3
h = 10. # layer thickness in km

# Non-Dimensionalized Parameters
P_star = rho* (R*1000.)**2 * Omega**2
rmin = (R-h)/R
rmax = R/R

physical_constants = {'R':R,'Omega':Omega,'rho':rho,'h':h,'P_star':P_star,'rmin':rmin,'rmax':rmax}
```

```python
# Size of Matricies
SizeM = 4*Nk*Nl+4*Nl

# Create model parameter vectors
dr = (rmax-rmin)/(Nk)
r = np.linspace(rmin-dr/2.,rmax+dr/2.,num=Nk+2.) # r value at center of each cell
rm = np.linspace(rmin-dr,rmax,num=Nk+2.) # r value at plus border (top) of cell
rp = np.linspace(rmin,rmax+dr,num=Nk+2.) # r value at minus border (bottom) of cell
dt = np.pi/(Nl)
t = np.linspace(-dt/2.,np.pi+dt/2.,num=Nl+2.) # theta value at center of cell
tm = np.linspace(-dt,np.pi,num=Nl+2.) # theta value at plus border (top) of cell
tp = np.linspace(0,np.pi+dt,num=Nl+2.) # theta value at minus border (bottom) of cell

model_parameters = {'Nk':Nk,'Nl':Nl,'SizeM':SizeM,
                    'dr':dr,'r':r,'rm':rm,'rp':rp, 
                    'dt':dt,'t':t,'tm':tm,'tp':tp,
                    'm_values':m_values,'l_values':l_values}
```

```python
### Create Model 
model = rf.Model(model_variables,physical_constants,model_parameters)
model.getIndex(1,1,'um')
model.add_gov_equation('rmom','uk')
# model.rmom.addTerm('p','-(rp[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=+1)
```

```python
###### Right Hand Matrix ########
def make_A(m):
    
    '''
    Creates the A matrix (M*l*x = A*x)
    m: azimuthal fourier mode to compute
    '''
    
    model.add_gov_equation('rmom','uk')
    model.add_gov_equation('tmom','ul')
    model.add_gov_equation('pmom','um')
    model.add_gov_equation('div','p')
    model.add_gov_equation('BC','p')

    model.rmom.addTerm('p','-(rp[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=+1)
    model.rmom.addTerm('p','+(rm[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=-1)
    model.rmom.addTerm('p','+(sin(tp[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=+1)
    model.rmom.addTerm('p','+(sin(tm[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=-1)
    model.rmom.addTerm('p','+(sin(tp[l])-sin(tm[l]))/(4.0*r[k]*sin(t[l]))',m)

    model.tmom.addTerm('p','-(sin(tp[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=+1)
    model.tmom.addTerm('p','+(sin(tm[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=-1)
    model.tmom.addTerm('p','+((sin(tm[l])-sin(tp[l]))/(2.0*dt) + cos(t[l]))/(r[k]*sin(t[l]))',m)
    model.tmom.addTerm('um','+2.0*cos(t[l])',m)

    model.pmom.addTerm('ul','-2.0*cos(t[l])',m)
    model.pmom.addTerm('p','-1j*m/(r[k]*sin(t[l]))',m)
    
    ### Div with all terms
    model.div.addTerm('uk','-(rp[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=+1)
    model.div.addTerm('uk','+(rm[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=-1)
    model.div.addTerm('uk','+(sin(tp[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=+1)
    model.div.addTerm('uk','+(sin(tm[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=-1)
    model.div.addTerm('uk','+(sin(tp[l])-sin(tm[l]))/(4.0*r[k]*sin(t[l]))',m)
    
    model.div.addTerm('ul','-(sin(tp[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=+1)
    model.div.addTerm('ul','+(sin(tm[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=-1)
    model.div.addTerm('ul','+((sin(tm[l])-sin(tp[l]))/(2.0*dt) + cos(t[l]))/(r[k]*sin(t[l]))',m)
    
    model.div.addTerm('um','-1j*m/(r[k]*sin(t[l]))',m)
    
    ### Div, taking out one term
#     model.div.addTerm('uk','-(rp[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=+1,first_l=2)
#     model.div.addTerm('uk','+(rm[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=-1,first_l=2)
#     model.div.addTerm('uk','+(sin(tp[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=+1,first_l=2)
#     model.div.addTerm('uk','+(sin(tm[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=-1,first_l=2)
#     model.div.addTerm('uk','+(sin(tp[l])-sin(tm[l]))/(4.0*r[k]*sin(t[l]))',m,first_l=2)
    
#     model.div.addTerm('ul','-(sin(tp[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=+1,first_l=2)
#     model.div.addTerm('ul','+(sin(tm[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=-1,first_l=2)
#     model.div.addTerm('ul','+((sin(tm[l])-sin(tp[l]))/(2.0*dt) + 1)/(r[k]*sin(t[l]))',m,first_l=2)
    
#     model.div.addTerm('um','-1j*m/(r[k]*sin(t[l]))',m,first_l=2)
    
#     model.div.addTerm('uk','-(rp[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=+1,first_k=2,last_l=1)
#     model.div.addTerm('uk','+(rm[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=-1,first_k=2,last_l=1)
#     model.div.addTerm('uk','+(sin(tp[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=+1,first_k=2,last_l=1,)
#     model.div.addTerm('uk','+(sin(tm[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=-1,first_k=2,last_l=1)
#     model.div.addTerm('uk','+(sin(tp[l])-sin(tm[l]))/(4.0*r[k]*sin(t[l]))',m,first_k=2,last_l=1)
    
#     model.div.addTerm('ul','-(sin(tp[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=+1,first_k=2,last_l=1)
#     model.div.addTerm('ul','+(sin(tm[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=-1,first_k=2,last_l=1)
#     model.div.addTerm('ul','+((sin(tm[l])-sin(tp[l]))/(2.0*dt) + 1)/(r[k]*sin(t[l]))',m,first_k=2,last_l=1)
    
#     model.div.addTerm('um','-1j*m/(r[k]*sin(t[l]))',m,first_k=2,last_l=1)
#     model.div.addValue('1',{'k':0,'l':1,'var':'p'},{'k':0,'l':1,'var':'p'})
    
    model.BC.addBC('uk','1.',0,m)
    model.BC.addBC('uk','1.',0,m,kdiff=1)
    model.BC.addBC('uk','1.',Nk+1,m)
    model.BC.addBC('uk','1.',Nk+1,m,kdiff=-1)

    model.BC.addBC('p','1.',0,m)
    model.BC.addBC('p','-1.',0,m,kdiff=1,first_l=1)
    model.BC.addBC('p','1.',Nk+1,m)
    model.BC.addBC('p','-1.',Nk+1,m,kdiff=-1)

    A = model.rmom.getCooMatrix() + model.tmom.getCooMatrix() \
    + model.pmom.getCooMatrix() + model.div.getCooMatrix() + model.BC.getCooMatrix()
    return A

####### Left Hand Matrix ########
def make_M(m):
    '''
    Creates the M matrix (M*l*x = A*x)
    m: azimuthal fourier mode to compute
    '''
    model.add_gov_equation('B_rmom','uk')
    model.add_gov_equation('B_tmom','ul')
    model.add_gov_equation('B_pmom','um')
    
    model.B_rmom.addTerm('uk','1',m)
    model.B_tmom.addTerm('ul','1',m)
    model.B_pmom.addTerm('um','1',m)
    M = model.B_rmom.getCooMatrix() + model.B_tmom.getCooMatrix() + model.B_pmom.getCooMatrix()
    return M
```

```python
### Make Matrices ###
A_matrices = {}
M_matrices = {}
for m in m_values:
    A_matrices[m] = make_A(m)
    M_matrices[m] = make_M(m)
```

```python
### Make starting vector for iteration ###
vec_0 = {}
for m in m_values:
    vec_0[m] = {}
    for l in l_values[m]:
        uk_0 = np.ones((Nk,Nl))*1.e-5
        ukBC_0 = np.ones((2,Nl))*1.e-5
        ul_0 = np.ones((Nk,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))
        um_0 = np.ones((Nk,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))
        p_0 = np.ones((Nk,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))
        pBC_0 = np.ones((2,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))

        vec_0[m][l] = model.create_vector(uk_0,ul_0,um_0,p_0,ukBC_0,pBC_0)
```

```python
##### Solve Eigenvalue Equation ######
eigenvectors = {}
eigenvalues = {}

# Eigenvalue to find
for m in m_values:
    eigenvectors[m] = {}
    eigenvalues[m] = {}
    for l in l_values[m]:
#         sigma_to_find = 2.*m/(l*(l+1.))*1j
        sigma_to_find = 0.58j
        num_eigs_to_find = 15
        tol = 1e-30
        eigenvalues[m][l],eigenvectors[m][l] = LA.eigs(A_matrices[m], k=num_eigs_to_find, M=M_matrices[m], 
                                                 sigma=sigma_to_find, return_eigenvectors=True, tol=tol)
        print ('sigma requested for m='+str(m)+', l='+str(l) 
               +' : ' + str(sigma_to_find) +', '+ str(len(eigenvalues[m][l]))+' eigenvalues found')
```

```python
##### Display 1D Waves ######
import matplotlib.cm as cm
colors = ['b','g','r','c']
for m in m_values:
    for l in l_values[m]:
        vecs = eigenvectors[m][l]
        vals = eigenvalues[m][l]
        for eignum in eigs_to_plot:
            fig, axes = plt.subplots(3,1,figsize=(15,8), sharex=True, sharey=True)
            vmin = min(vecs[:,eignum]).real
            vmax = max(vecs[:,eignum]).real
            titles = ['Absolute Value','Real','Imaginary']
            for (ax,title,ind) in zip(axes,titles,range(len(axes))):
                ax.set_title(title)
                line = []
                for (var,color) in zip(model.model_variables,colors):
                    if ind == 0:
                        line.append(ax.plot(t[1:-1]*180./np.pi,abs(model.getVariable(vecs[:,eignum],var).T),color=color))
                        ax.legend([x[0] for x in line],model.model_variables,loc=0)
                    elif ind ==1:
                        line.append(ax.plot(t[1:-1]*180./np.pi,model.getVariable(vecs[:,eignum],var).T.real,color=color))
                    else:
                        line.append(ax.plot(t[1:-1]*180./np.pi,model.getVariable(vecs[:,eignum],var).T.imag,color=color))
                ax.grid()
            plt.suptitle('Eigenvalue = '+str(vals[eignum])+', m='+str(m)+
                             ' Nk='+str(Nk)+' Nl='+str(Nl), size=16)
```
