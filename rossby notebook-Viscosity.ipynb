Solves the system of equations 
$$\frac{d\vec{u}}{dt}=-\frac{1}{\rho} \nabla P - 2\Omega \times \vec{u} + \nu \nabla ^2 \vec{u}$$
$$ \nabla \cdot \vec{u} = 0 $$ 
Nk points in radial direction <br>
Nl points in theta direction

```python
import numpy as np
import scipy as sp
import scipy.sparse as sparse
import scipy.sparse.linalg as LA
import matplotlib.pyplot as plt
import matplotlib.pylab as pyl
import matplotlib as mpl
import rossby_functions_viscosity as rf

from numpy import sin
from numpy import cos
%matplotlib inline
reload(rf)
```

```python
# modes to simulate
m_min = 2
m_max = 2
m_values = range(m_min,m_max+1)
l_values = {}
l_values[m_values[0]]=[m_values[0]]

# Size of grid
Nk = 5
Nl = 201
Nm = 1

model_variables = ('uk','ul','um','p')

# Physical Constants
R = 3160. # Outer core radius in km
Omega = np.pi/(24.0*3600.0) # rotation rate in rad/s
rho = 1.e5  # density in kg/m^3
h = 10. # layer thickness in km
nu = 1e2 # kinematic viscosity in m^2/s

# Non-Dimensionalized Parameters
P_star = rho* (R*1000.)**2 * Omega**2
E = nu/(Omega * (R*1000.)**2)
rmin = (R-h)/R
rmax = R/R

print E
physical_constants = {'R':R,'Omega':Omega,'rho':rho,'h':h,'P_star':P_star,'rmin':rmin,'rmax':rmax,'E':E,'nu':nu}
```

```python
# Size of Matricies
SizeM = 4*Nk*Nl+8*Nl

# Create model parameter vectors
dr = (rmax-rmin)/(Nk)
r = np.linspace(rmin-dr/2.,rmax+dr/2.,num=Nk+2.) # r value at center of each cell
rm = np.linspace(rmin-dr,rmax,num=Nk+2.) # r value at plus border (top) of cell
rp = np.linspace(rmin,rmax+dr,num=Nk+2.) # r value at minus border (bottom) of cell
dt = np.pi/(Nl)
t = np.linspace(-dt/2.,np.pi+dt/2.,num=Nl+2.) # theta value at center of cell
tm = np.linspace(-dt,np.pi,num=Nl+2.) # theta value at plus border (top) of cell
tp = np.linspace(0,np.pi+dt,num=Nl+2.) # theta value at minus border (bottom) of cell

model_parameters = {'Nk':Nk,'Nl':Nl,'SizeM':SizeM,
                    'dr':dr,'r':r,'rm':rm,'rp':rp, 
                    'dt':dt,'t':t,'tm':tm,'tp':tp,
                    'm_values':m_values,'l_values':l_values}
```

```python
### Create Model 
model = rf.Model(model_variables,physical_constants,model_parameters)
# print model.getIndex(1,1,'uk')
# print SizeM
# model.add_gov_equation('rmom','uk')
# model.rmom.addTerm('p','-(rp[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=+1)
```

```python
###### Right Hand Matrix ########
def make_A(m):
    
    '''
    Creates the A matrix (M*l*x = A*x)
    m: azimuthal fourier mode to compute
    '''
    
    model.add_gov_equation('rmom','uk')
    model.add_gov_equation('tmom','ul')
    model.add_gov_equation('pmom','um')
    model.add_gov_equation('div','p')
    model.add_gov_equation('BC','p')
    
    ### R-momentum
    model.rmom.addTerm('p','-(rp[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=+1)
    model.rmom.addTerm('p','+(rm[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=-1)
    model.rmom.addTerm('p','+(sin(tp[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=+1)
    model.rmom.addTerm('p','+(sin(tm[l])/sin(t[l]))/(4.0*r[k])',m,ldiff=-1)
    model.rmom.addTerm('p','+(sin(tp[l])-sin(tm[l]))/(4.0*r[k]*sin(t[l]))',m)
    model.rmom.addTerm('uk','+E/dr**2',m,kdiff=+1)
    model.rmom.addTerm('uk','-2*E/dr**2',m)
    model.rmom.addTerm('uk','+E/dr**2',m,kdiff=-1)

    ### Theta-Momentum
    model.tmom.addTerm('p','-(sin(tp[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=+1)
    model.tmom.addTerm('p','+(sin(tm[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=-1)
    model.tmom.addTerm('p','+((sin(tm[l])-sin(tp[l]))/(2.0*dt) + cos(t[l]))/(r[k]*sin(t[l]))',m)
    model.tmom.addTerm('um','+2.0*cos(t[l])',m)
    model.tmom.addTerm('ul','+E/dr**2',m,kdiff=+1)
    model.tmom.addTerm('ul','-2*E/dr**2',m)
    model.tmom.addTerm('ul','+E/dr**2',m,kdiff=-1)
    
    ### Phi-Momentum
    model.pmom.addTerm('ul','-2.0*cos(t[l])',m)
    model.pmom.addTerm('p','-1j*m/(r[k]*sin(t[l]))',m)
    model.pmom.addTerm('um','+E/dr**2',m,kdiff=+1)
    model.pmom.addTerm('um','-2*E/dr**2',m)
    model.pmom.addTerm('um','+E/dr**2',m,kdiff=-1)
    
    ### Divergence (Mass Conservation)
    model.div.addTerm('uk','+(rp[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=+1)
    model.div.addTerm('uk','-(rm[k]/r[k])**2.0 / (2.0*dr)',m,kdiff=-1)
    model.div.addTerm('uk','+1/r[k]',m)
    model.div.addTerm('ul','+(sin(tp[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=+1)
    model.div.addTerm('ul','-(sin(tm[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=-1)
    model.div.addTerm('ul','+(sin(tp[l])-sin(tm[l]))/(2.0*dt*r[k]*sin(t[l]))',m)
    model.div.addTerm('um','1j*m/(r[k]*sin(t[l]))',m)
    
    ### Boundary Conditions
    model.BC.addBC('uk','1.',0,m)
    model.BC.addBC('uk','1.',0,m,kdiff=1)
    model.BC.addBC('uk','1.',Nk+1,m)
    model.BC.addBC('uk','1.',Nk+1,m,kdiff=-1)

    model.BC.addBC('ul','1.',0,m)
    model.BC.addBC('ul','-1.',0,m,kdiff=1)
    model.BC.addBC('ul','1.',Nk+1,m)
    model.BC.addBC('ul','-1.',Nk+1,m,kdiff=-1)

    model.BC.addBC('um','1.',0,m)
    model.BC.addBC('um','-1.',0,m,kdiff=1)
    model.BC.addBC('um','1.',Nk+1,m)
    model.BC.addBC('um','-1.',Nk+1,m,kdiff=-1)
    
    model.BC.addBC('p','1.',0,m)
    model.BC.addBC('p','-1.',0,m,kdiff=1,first_l=1)
    model.BC.addBC('p','1.',Nk+1,m)
    model.BC.addBC('p','-1.',Nk+1,m,kdiff=-1)
    
    A = model.rmom.getCooMatrix() + model.tmom.getCooMatrix() \
    + model.pmom.getCooMatrix() + model.div.getCooMatrix() + model.BC.getCooMatrix()
    return A

####### Left Hand Matrix ########
def make_M(m):
    '''
    Creates the M matrix (M*l*x = A*x)
    m: azimuthal fourier mode to compute
    '''
    model.add_gov_equation('B_rmom','uk')
    model.add_gov_equation('B_tmom','ul')
    model.add_gov_equation('B_pmom','um')
    
    model.B_rmom.addTerm('uk','1',m)
    model.B_tmom.addTerm('ul','1',m)
    model.B_pmom.addTerm('um','1',m)
    M = model.B_rmom.getCooMatrix() + model.B_tmom.getCooMatrix() + model.B_pmom.getCooMatrix()
    return M
```

```python
### Make Matrices ###
A_matrices = {}
M_matrices = {}
for m in m_values:
    A_matrices[m] = make_A(m)
    M_matrices[m] = make_M(m)
```

```python
### Make starting vector for iteration ###
vec_0 = {}
for m in m_values:
    vec_0[m] = {}
    for l in l_values[m]:
        uk_0 = np.ones((Nk,Nl))*1.e-5
        ukBC_0 = np.ones((2,Nl))*1.e-5
        ul_0 = np.ones((Nk,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))
        ulBC_0 = np.ones((2,Nl))*1.e-5
        um_0 = np.ones((Nk,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))
        umBC_0 = np.ones((2,Nl))*1.e-5
        p_0 = np.ones((Nk,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))
        pBC_0 = np.ones((2,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))

        vec_0[m][l] = model.create_vector(uk_0,ul_0,um_0,p_0,ukBC_0,ulBC_0,umBC_0,pBC_0)
```

```python
##### Solve Eigenvalue Equation ######
eigenvectors = {}
eigenvalues = {}
# sigma_list = [0.586j, 0.196j, 0.09j, 0.0501j, 0.0315j, 0.0215j, 0.0155j]
sigma_list = [0.3333j]
sigmas = sigma_list

# Eigenvalue to find
for m in m_values:
    eigenvectors[m] = {}
    eigenvalues[m] = {}
    for l in l_values[m]:
        for (sigma_to_find,ind) in zip(sigmas,range(len(sigmas))):
            num_eigs_to_find = 4
            tol = 1e-30
            eigenvalues[m][l+ind],eigenvectors[m][l+ind] = LA.eigs(A_matrices[m], k=num_eigs_to_find, M=M_matrices[m], 
                                                     sigma=sigma_to_find, return_eigenvectors=True, tol=tol)
            print ('sigma requested for m='+str(m)+', l='+str(l+ind) 
                   +' : ' + str(sigma_to_find) +', '+ str(len(eigenvalues[m][l]))+' eigenvalues found')
```

```python
### Spherical Harmonics Wave Equation with vorticity
def u_m2l2_a(th,ph):
    C = 0.045*(0.25*np.sqrt(15.0/(2.*np.pi)))
    u_th = 2j*C*np.exp(2j*ph)*sin(th)
    u_ph = -2*C*np.exp(2j*ph)*sin(th)*cos(th)
    return u_th, u_ph

def u_m2l2_n(th,ph):
    u_th = 0.036*sin(th)**2
    u_ph = 0.053j*sin(th)**2*cos(th)
    return u_th, u_ph

def u_m2l3_a(th,ph):
    C = 0.03*0.25*np.sqrt(105.0/(2*np.pi))
    u_th = 2j*C*np.exp(2j*ph)*sin(th)*cos(th)
    u_ph = -0.5*C*np.exp(2j*ph)*sin(th)*(3*cos(2*th)+1)
    return u_th, u_ph

def u_m2l3_n(th,ph):
    u_th = 0.065*(1-1j)*sin(th)**4*cos(th)
    u_ph = 0.015*(1+1j)*sin(th)**3*(5*cos(2*th)+3)
    return u_th, u_ph

def u_m2l4_a(th,ph):
    C = 0.022*3.0/8.0*np.sqrt(5.0/(2*np.pi))
    u_th = 2j*C*np.exp(2j*ph)*(sin(th))*(7*(cos(th))**2-1)
    u_ph = -3./2.*C*np.exp(2j*ph)*sin(2*th)*(7*cos(2*th)-1)
    return u_th, u_ph

def u_m2l4_n(th,ph):
    u_th = 0.012*(1.25-0.5j)*(sin(th))**6*(16*(cos(th))**2-1)
    u_ph = 0.005*(0.5+1.35j)*2*sin(th)**5*cos(th)*(32*cos(2*th)+13)
    return u_th, u_ph

u2th_a,u2ph_a = u_m2l2_a(t[1:-1],np.pi*3./4.)
u2th_n,u2ph_n = u_m2l2_n(t[1:-1],0.)
u3th_a,u3ph_a = u_m2l3_a(t[1:-1],np.pi*4./6.)
u3th_n,u3ph_n = u_m2l3_n(t[1:-1],0.)
u4th_a,u4ph_a = u_m2l4_a(t[1:-1],np.pi*2./3.)
u4th_n,u4ph_n = u_m2l4_n(t[1:-1],0.)

plt.figure()
plt.suptitle('l=2')
plt.subplot(2,1,1)
plt.plot(t[1:-1]*180./np.pi,u2th_a.real,'g:')
plt.plot(t[1:-1]*180./np.pi,u2ph_a.real,'r:')
plt.plot(t[1:-1]*180./np.pi,u2th_n.real,'g--')
plt.plot(t[1:-1]*180./np.pi,u2ph_n.real,'r--')
plt.grid()
plt.subplot(2,1,2)
plt.plot(t[1:-1]*180./np.pi,u2th_a.imag,'g:')
plt.plot(t[1:-1]*180./np.pi,u2ph_a.imag,'r:')
plt.plot(t[1:-1]*180./np.pi,u2th_n.imag,'g--')
plt.plot(t[1:-1]*180./np.pi,u2ph_n.imag,'r--')
plt.grid()

plt.figure()
plt.suptitle('l=3')
plt.subplot(2,1,1)
plt.grid()
plt.plot(t[1:-1]*180./np.pi,u3th_a.real,'g:')
plt.plot(t[1:-1]*180./np.pi,u3ph_a.real,'r:')
plt.plot(t[1:-1]*180./np.pi,u3th_n.real,'g--')
plt.plot(t[1:-1]*180./np.pi,u3ph_n.real,'r--')
plt.subplot(2,1,2)
plt.plot(t[1:-1]*180./np.pi,u3th_a.imag,'g:')
plt.plot(t[1:-1]*180./np.pi,u3ph_a.imag,'r:')
plt.plot(t[1:-1]*180./np.pi,u3th_n.imag,'g--')
plt.plot(t[1:-1]*180./np.pi,u3ph_n.imag,'r--')
plt.grid()

plt.figure()
plt.suptitle('l=4')
plt.subplot(2,1,1)
plt.plot(t[1:-1]*180./np.pi,u4th_a.real,'g:')
plt.plot(t[1:-1]*180./np.pi,u4th_n.real,'g--')
plt.plot(t[1:-1]*180./np.pi,u4ph_n.real,'r--')
plt.plot(t[1:-1]*180./np.pi,u4ph_a.real,'r:')
plt.grid()
plt.subplot(2,1,2)
plt.plot(t[1:-1]*180./np.pi,u4th_a.imag,'g:')
plt.plot(t[1:-1]*180./np.pi,u4ph_a.imag,'r:')
plt.plot(t[1:-1]*180./np.pi,u4th_n.imag,'g--')
plt.plot(t[1:-1]*180./np.pi,u4ph_n.imag,'r--')

```
