Solves the system of equations 
$$\frac{d\vec{u}}{dt}=-\nabla P + 2\Omega \times \vec{u}$$
$$ \nabla \cdot \vec{u} = 0 $$ 

```python
import numpy as np
import scipy as sp
import scipy.sparse as sparse
import scipy.sparse.linalg as LA
import matplotlib.pyplot as plt
import matplotlib.pylab as pyl
import matplotlib as mpl
from numpy import sin
from numpy import cos
%matplotlib inline
```

```python
# modes to simulate
m_min = 2
m_max = 2
m_values = range(m_min,m_max+1)
l_values = {}
l_values[m_values[0]]=[m_values[0]]


# Size of grid
Nk = 3
Nl = 30
Nm = 1

ModelVariables = ('uk','ul','um','p')

# Physical Parameters
R = 3160. # Outer core radius in km
Omega = np.pi/(24.0*3600.0) # rotation rate in rad/s
rho = 1.e5  # density in kg/m^3
h = 10. # layer thickness in km

# Non-Dimensionalized Parameters
P_star = rho* (R*1000.)**2 * Omega**2
rmin = (R-h)/R
rmax = R/R

print rmin, rmax, m_values
print 'Omega = '+str(Omega) +' rad/s'
print 'u_star = '+str(Omega*R*1000) +' m/s'
print 'P_star = '+str(P_star/1e6)+' GPa'
```

```python
# Size of Matricies
Sizeuk = Nk*Nl
Sizeul = Nk*Nl
Sizeum = Nk*Nl
Sizep = Nk*Nl
SizeukBC = 2*Nl
SizepBC = 2*Nl
SizeMnoBC = Sizeuk+Sizeul+Sizeum+Sizep
SizeMwithBC = SizeMnoBC+SizeukBC+SizepBC

# Create model parameter vectors
dr = (rmax-rmin)/(Nk)
r = np.linspace(rmin-dr/2.,rmax+dr/2.,num=Nk+2.) # r value at center of each cell
rm = np.linspace(rmin-dr,rmax,num=Nk+2.) # r value at plus border (top) of cell
rp = np.linspace(rmin,rmax+dr,num=Nk+2.) # r value at minus border (bottom) of cell
dt = np.pi/(Nl)
t = np.linspace(-dt/2.,np.pi+dt/2.,num=Nl+2.) # theta value at center of cell
tm = np.linspace(-dt,np.pi,num=Nl+2.) # theta value at plus border (top) of cell
tp = np.linspace(0,np.pi+dt,num=Nl+2.) # theta value at minus border (bottom) of cell
```

```python
##### Functions to Use ####
def getIndex(k,l,var):
    '''
    Takes coordinates for a point, gives back index in matrix.
    :param k: k grid value from 1 to K (or 0 to K+1 for uk or p)
    :param l: l grid value from 1 to L
    :param var: 'uk', 'ul', 'um', or 'p'
    :return: index of location in matrix
    '''
    if (var in ModelVariables):
        if var == 'uk':
            if l>=1 and l<=Nl:
                # Check for boundary point request, assign at end of matrix
                if k==0:
                    return  SizeMnoBC + l-1
                elif k==Nk+1:
                    return SizeMnoBC + Nl + l-1
                elif k>=1 and k<=Nk:
                    return  (k-1) + (l-1)*Nk
                else:
                    return 'k index out of bounds'
            else:
                return ' l index out of bounds'
            
        elif var == 'ul':
            if k>=1 and k<=Nk and l>=0 and l<=Nl+1:
                return Sizeuk + k-1 + (l-1)*Nk
            else:
                return 'k or l index out of bounds'

        elif var == 'um':
            if k>=1 and k<=Nk and l>=1 and l<=Nl:
                return Sizeuk + Sizeul + k-1 + (l-1)*Nk
            else:
                return 'k or l index out of bounds'
            
        elif var == 'p':
            if l>=1 and l<=Nl:
                # Check for boundary point request, assign at end of matrix
                if k==0:
                    return  SizeMnoBC + SizeukBC + l-1
                elif k==Nk+1:
                    return SizeMnoBC + SizeukBC + Nl + l-1
                elif k>=1 and k<=Nk:
                    return  Sizeuk + Sizeul + Sizeum + k-1 + (l-1)*Nk
                else:
                    return 'k index out of bounds'
            else:
                return 'l or m index out of bounds'
    else:
        return 'Not a valid variable name in this model'

def getVariable(vector,var):
    if (var in ModelVariables):
        if var == 'uk':
            return np.reshape(vector[0:Sizeuk],(Nk,Nl))

        elif var == 'ul':
            return np.reshape(vector[Sizeuk:Sizeuk+Sizeul],(Nk,Nl))

        elif var == 'um':
            return np.reshape(vector[Sizeuk+Sizeul:Sizeuk+Sizeul+Sizeum],(Nk,Nl))

        elif var == 'p':
            return np.reshape(vector[Sizeuk+Sizeul+Sizeum:SizeMnoBC],(Nk,Nl))
    elif var == 'ukBC':
        return np.reshape(vector[SizeMnoBC:SizeMnoBC+Nl*2],(2,Nl))
    elif var == 'pBC':
        return np.reshape(vector[SizeMnoBC+Nl*2:SizeMwithBC],(2,Nl))
    else:
        return 'Not a valid variable name in this model'

class GovEquation():
    def __init__(self,variable):
        self.rows = []
        self.cols = []
        self.vals = []
        self.variable = variable

    def addTerm(self,var,value,m,kdiff=0,ldiff=0,mdiff=0):
        for l in range(1,Nl+1):
            for k in range(1,Nk+1):
                value_computed = eval(value,globals(),locals())
                if not np.isclose(value_computed,0.0,atol=1E-12,rtol=1E-12):
                    if var =='uk':
                        if (k >= 1 and k <=Nk) and (l >= 1 and l <=Nl):
                            self.rows.append(getIndex(k,l,self.variable))
                            self.cols.append(getIndex(k+kdiff,l+ldiff,var))
                            self.vals.append(value_computed)
                        else:
                            return 'index out of bounds'
                    elif var =='ul':
                        if (k >= 1 and k <=Nk) and (l >= 1 and l <=Nl):
                            self.rows.append(getIndex(k,l,self.variable))
                            self.cols.append(getIndex(k+kdiff,l+ldiff,var))
                            self.vals.append(value_computed)
                        else:
                            return 'index out of bounds'
                    elif var =='um':
                        if (k >= 1 and k <=Nk) and (l >= 1 and l <=Nl):
                            self.rows.append(getIndex(k,l,self.variable))
                            self.cols.append(getIndex(k+kdiff,l+ldiff,var))
                            self.vals.append(value_computed)
                        else:
                            return 'index out of bounds'
                    elif var =='p':
                        if (k >= 1 and k <=Nk) and (l >= 1 and l <=Nl):
                            self.rows.append(getIndex(k,l,self.variable))
                            self.cols.append(getIndex(k+kdiff,l+ldiff,var))
                            self.vals.append(value_computed)
                        else:
                            return 'index out of bounds'

    def addBC(self,var,value,k,m,kdiff=0,first=1,last=Nl):
        if var == 'uk':
            for l in range(first,last+1):
                self.rows.append(getIndex(k,l,var))
                self.cols.append(getIndex(k+kdiff,l,var))
                self.vals.append(eval(value,globals(),locals()))
        elif var == 'p':
            for l in range(first,last+1):
                self.rows.append(getIndex(k,l,var))
                self.cols.append(getIndex(k+kdiff,l,var))
                self.vals.append(eval(value,globals(),locals()))

    def addValue(self,value,row,col):
        '''
        Adds a term to a specific index.
        value: term to add
        row: dictionary containing 'k','l',and 'var'
        col: dictionary containing 'k','l',and 'var'
        '''
        self.rows.append(getIndex(row[k],row[l],row[var]))
        self.cols.append(getIndex(col[k],col[l],col[var]))
        self.vals.append(eval(value,globals(),locals()))
        
    def getCooMatrix(self):
        return sparse.coo_matrix((self.vals,(self.rows,self.cols)),shape=(SizeMwithBC,SizeMwithBC))

    def todense(self):
        return self.getCooMatrix().todense()

    def getCsrMatrix(self):
        return self.getCooMatrix().tocsr()
    
###### Right Hand Matrix ########
def make_A(m):
    '''
    Creates the A matrix (M*l*x = A*x)
    m: azimuthal fourier mode to compute
    '''
    rmom = GovEquation('uk')
    tmom = GovEquation('ul')
    pmom = GovEquation('um')
    div =GovEquation('p')
    BC = GovEquation('p')

    rmom.addTerm('p','-1/(2.0*dr)',m,kdiff=+1)
    rmom.addTerm('p','+1/(2.0*dr)',m,kdiff=-1)

    tmom.addTerm('p','-(sin(tp[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=+1)
    tmom.addTerm('p','+(sin(tm[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=-1)
    tmom.addTerm('um','+2.0*cos(t[l])',m)

    pmom.addTerm('ul','-2.0*cos(t[l])',m)
    pmom.addTerm('p','-1j*m/(r[k]*sin(t[l]))',m)

    div.addTerm('uk','-1 / (2.0*dr)',m,kdiff=+1)
    div.addTerm('uk','+1 / (2.0*dr)',m,kdiff=-1)
    
    div.addTerm('ul','-(sin(tp[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=+1)
    div.addTerm('ul','+(sin(tm[l])/sin(t[l]))/(2.0*r[k]*dt)',m,ldiff=-1)
    
    div.addTerm('um','-1j*m/(r[k]*sin(t[l]))',m)

    ### uk = 0 at r_top and r_bottom
    BC.addBC('uk','1.',0,m)
    BC.addBC('uk','1.',0,m,kdiff=1)
    BC.addBC('uk','1.',Nk+1,m)
    BC.addBC('uk','1.',Nk+1,m,kdiff=-1)

    ### dP/dr = 0 at r_top and r_bottom
    BC.addBC('p','1.',0,m)
    BC.addBC('p','-1.',0,m,kdiff=1,first=2) 
    ''' By starting at second cell at theta=0 (first=2), the cell nearest N. pole has boundary 
    condition of P(k+1)=0 rather than dP/dr = 0 at boundary '''
    BC.addBC('p','1.',Nk+1,m)
    BC.addBC('p','-1.',Nk+1,m,kdiff=-1)
    
    A = rmom.getCooMatrix() + tmom.getCooMatrix() + pmom.getCooMatrix() + div.getCooMatrix() + BC.getCooMatrix()
    return A

####### Left Hand Matrix ########
def make_M(m):
    '''
    Creates the M matrix (M*l*x = A*x)
    m: azimuthal fourier mode to compute
    '''
    lrmom = GovEquation('uk')
    ltmom = GovEquation('ul')
    lpmom = GovEquation('um')
    lrmom.addTerm('uk','1',m)
    ltmom.addTerm('ul','1',m)
    lpmom.addTerm('um','1',m)
    M = lrmom.getCooMatrix() + ltmom.getCooMatrix() + lpmom.getCooMatrix()
    return M

def create_vector(uk,ul,um,p,ukBC,pBC):
    return np.concatenate((np.reshape(uk,(Nk*Nl,1)),np.reshape(ul,(Nk*Nl,1)),np.reshape(um,(Nk*Nl,1)),np.reshape(p,(Nk*Nl,1)),
                           np.reshape(ukBC,(2*Nl,1)), np.reshape(pBC,(2*Nl,1))), axis=0)
```

```python
### Make Matrices ###
A_matrices = {}
M_matrices = {}
for m in m_values:
    A_matrices[m] = make_A(m)
    M_matrices[m] = make_M(m)
```

```python
### Make starting vector for iteration ###
vec_0 = {}
for m in m_values:
    vec_0[m] = {}
    for l in l_values[m]:
        uk_0 = np.ones((Nk,Nl))*1.e-5
        ukBC_0 = np.ones((2,Nl))*1.e-5
        ul_0 = np.ones((Nk,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))
        um_0 = np.ones((Nk,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))
        p_0 = np.ones((Nk,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))
        pBC_0 = np.ones((2,1))*np.matrix(sp.special.sph_harm(m,l,0.*np.ones_like(t[1:-1]),t[1:-1]))

        vec_0[m][l] = create_vector(uk_0,ul_0,um_0,p_0,ukBC_0,pBC_0)
```

```python
##### Solve Eigenvalue Equation ######
eigenvectors = {}
eigenvalues = {}

# Eigenvalue to find
for m in m_values:
    eigenvectors[m] = {}
    eigenvalues[m] = {}
    for l in l_values[m]:
#         sigma_to_find = 2.*m/(l*(l+1.))
        sigma_to_find = 1e-2
        num_eigs_to_find = 10
        tol = 1e-30
        eigenvalues[m][l],eigenvectors[m][l] = LA.eigs(A_matrices[m], k=num_eigs_to_find, M=M_matrices[m], 
                                                 sigma=sigma_to_find, return_eigenvectors=True, tol=tol)
        print ('sigma requested for m='+str(m)+', l='+str(l) 
               +' : ' + str(sigma_to_find) +', '+ str(len(eigenvalues[m][l]))+' eigenvalues found')
```

```python
# Finds the eigenvectors to plot
# plot = 'SI'
# plot = 'all'
plot = 2

if plot == 'SI':
    for m in m_values:
        for l in l_values[m]:
            imag_max = np.amax(abs(eigenvectors[m][l].imag),axis=0)
            real_max = np.amax(abs(eigenvectors[m][l].real),axis=0)
            eigs_to_plot = [np.argmin(imag_max-real_max,axis=0)]
elif plot == 'all':
    eigs_to_plot = range(num_eigs_to_find)
else: 
    eigs_to_plot = range(plot)
print eigs_to_plot
print eigenvalues[m][l][eigs_to_plot]
```

```python
##### Display 2D Waves ######
import matplotlib.cm as cm
for m in m_values:
    for l in l_values[m]:
        vecs = eigenvectors[m][l]
        vals = eigenvalues[m][l]
        for eignum in eigs_to_plot:
            fig, axes = plt.subplots(4,2,figsize=(15,8), sharex=True, sharey=True)
            vmin = min(vecs[:,eignum]).real
            vmax = max(vecs[:,eignum]).real
            for (var,ax) in zip(ModelVariables,axes):
                if var == 'p':
                    Z = getVariable(vecs[:,eignum],var).real
                    ax[0].set_title(var+' real (Pa)')
                    ax[0].imshow(Z, cmap=cm.gray, interpolation='none', extent=[0,180,rmin*R,rmax*R],aspect='auto', vmin=vmin,vmax=vmax )
                    ax[0].set_xlabel('theta (degrees)')
                    ax[0].grid()

                    Z_im = getVariable(vecs[:,eignum],var).imag
                    ax[1].set_title(var+' imag (Pa)')
                    ax[1].imshow(Z_im, cmap=cm.gray, interpolation='none', extent=[0,180,rmin*R,rmax*R],aspect='auto', vmin=vmin,vmax=vmax  )
                    ax[1].set_xlabel('theta (degrees)')
                    ax[1].grid()
                else:
                    Z = getVariable(vecs[:,eignum],var).real
                    ax[0].set_title(var+' real (m/s)')
                    im0 = ax[0].imshow(Z, cmap=cm.RdBu, interpolation='none', extent=[0,180,rmin*R,rmax*R],aspect='auto', vmin=vmin,vmax=vmax  )
                    ax[0].grid()

                    Z_im = getVariable(vecs[:,eignum],var).imag
                    ax[1].set_title(var+' imag (m/s)')
                    im1 = ax[1].imshow(Z_im, cmap=cm.RdBu, interpolation='none', extent=[0,180,rmin*R,rmax*R],aspect='auto', vmin=vmin,vmax=vmax )
                    ax[1].grid()
                ax[0].set_ylabel('R (km)')
                plt.suptitle('Eigenvalue = '+str(vals[eignum])+', l='+str(l)+' m='+str(m)+
                             ' Nk='+str(Nk)+' Nl='+str(Nl), size=16)
                plt.colorbar(im0)

```

```python
### Plot A Matrix (M*l*x = A*x) ###

for m in m_values:
    A = A_matrices[m]
    plt.figure(figsize=(15,20))
    plt.subplot(3,2,1)
    plt.spy(np.abs(A.todense()))
    plt.grid()
    plt.title('A'+str(m)+' matrix all terms (M*l*x = A*x)')

    plt.subplot(3,2,2)
    plt.spy(np.abs(A.todense().real))
    plt.grid()
    plt.title('A'+str(m)+' real')

    plt.subplot(3,2,3)
    plt.spy(np.abs(A.todense().imag))
    plt.grid()
    plt.title('A'+str(m)+' imaginary')

    A_pos = np.matrix(A.todense())
    A_pos[A.todense()<0.] = 0
    plt.subplot(3,2,4)
    plt.spy(np.abs(A_pos))
    plt.grid()
    plt.title('A'+str(m)+' positive')

    A_neg = np.matrix(A.todense())
    A_neg[A.todense()>0.] = 0
    plt.subplot(3,2,5)
    plt.spy(np.abs(A_neg))
    plt.grid()
    plt.title('A'+str(m)+' negative')
    plt.tight_layout()
```
